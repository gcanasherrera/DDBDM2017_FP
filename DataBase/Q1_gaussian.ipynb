### METHOD 1: Calculation of Kernel Density


```python
# Create object from sklearn KernelDensity
from sklearn.neighbors import KernelDensity
kde = KernelDensity(bandwidth=bw_optimal, kernel='gaussian').fit(both)
```

```python
kde = KernelDensity(bandwidth=0.0436, kernel='gaussian').fit(both)

#All of this is needed for the countour!
xmin = np.min(Color.colorYJ.values)
xmax = np.max(Color.colorYJ.values)
ymin = np.min(Color.colorJH.values)
ymax = np.max(Color.colorJH.values)
xgrid = np.linspace(xmin, xmax, 100)
ygrid = np.linspace(ymin, ymax, 100)
X, Y = np.meshgrid(xgrid,ygrid)
XX = np.array([X.ravel(),Y.ravel()]).T
kde_density_log = kde.score_samples(XX)
kde_density = np.exp(kde_density_log)
kde_density = kde_density.reshape(X.shape)
num_levels = 30
levels = np.logspace(-2,2,num_levels)
cmap = plt.cm.get_cmap()
colors = cmap(levels)
colors='grey'

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.plot(Color.colorYJ.values, Color.colorJH.values, linestyle='none', marker='.', label="Stars")
ax.contour(X, Y, kde_density, levels=levels,colors=colors, label="KDE Reconstruction")

ax.set_xlabel('colorYJ', fontsize=15)
ax.set_ylabel('colorJH', fontsize=15)
```

### METHOD 2: GaussianMixture from sklearn
Let's try to reproduce the results with sklearn mixture model. First, we need to obtain an estimation of the number of gaussians we need.

```python
#Let's import sklearn mixture model as well as stacking all data to fit
from sklearn import mixture 
X = np.vstack([Color.colorYJ.values, Color.colorJH.values]).T
```

```python
#Let's calculate the BIC and AIC for gaussian models regarding different mix number
#According to Juan (linkedin.com/in/jclaramunt), usually, 
#BIC and AIC are positive: problem with the normalization. Still, we can trust the values "more negative".
number_gaussian=np.arange(1, 15, 1)
bic=[]
aic=[]
likelihood=[]
for element in number_gaussian:
    model= mixture.GaussianMixture(element)
    model.fit(X)
    bic.append(model.bic(X))
    aic.append(model.aic(X))
    likelihood.append(model._estimate_log_prob(X))
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(number_gaussian, bic, linestyle='dashed', marker='o', label="BIC")
ax.plot(number_gaussian, aic,linestyle='solid', marker='^',  label="AIC")
ax.set_xlabel("Number of Gaussian functions", fontsize=15)
ax.set_ylabel("Information Criteria", fontsize=15)
#ax.set_yscale("log")
#ax.set_xscale("log")
ax.legend(fontsize=15)
ax.set_xlim(2, 15)
#Detailed to zoom up!
ax.set_ylim(-56000, -53000)
```

```python
#Let's choose 6 as the number of gaussian functions and create the model
model_mix = mixture.GaussianMixture(6)
X = np.vstack([Color.colorYJ.values, Color.colorJH.values]).T
model_mix.fit(X)
sim_stars = model_mix.sample(100000) #careful, the information is saved in the first component
#print sim_stars[0][:,0]
```

### METHOD: BayesianGaussianMixture

Check with BayesianGaussianMixture to see whether the number of components = 6 was a good estimation. This method gives 13 instead... does this provide some discussion? Probably, but 13 means that it is fitting the stars in the tails of the distribution.

```python
Xbayesian = np.vstack([Color.colorYJ.values, Color.colorJH.values]).T
model_mix_bayesian = mixture.BayesianGaussianMixture(n_components=20, max_iter=5000)
model_mix_bayesian.fit(Xbayesian)
def_n_comp=model_mix_bayesian.predict(Xbayesian)
print "Number of components estimated by Bayesian Gaussian Mixture: {0}".format(len(np.unique(def_n_comp)))
```

```python
sim_stars_bayesian = model_mix_bayesian.sample(100000)
```
