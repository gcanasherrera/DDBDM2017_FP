```python
def do_regression(model, M, y, modelst='none', alpha=0.):
    """
    This function fits x-data M and y-data y using 'model' as a regression model. 
    It calculates the value for the error E_theta as defined in the problemset. 
    
    """
    
    # the higher z, the less accurate: less weight on higher zs:
    #weights = 1./y # this does decrease the linear error a tiny tiny bit...
    # the higher magr, the less accurate: less weight on higher mags:
    #weights = 1./(10**(-2.5*data_train['mag_r']))
    #weights = y/(1+y)
    
    # Fitting the model
    #res = model.fit(M,y, sample_weight=weights)
    res = model.fit(M,y)
    # Calculating the predicted values of y given the model (with parameters theta) for given data M
    ypred = model.predict(M)
    # Calculating the error from the problem set
    E_theta = np.median( np.abs( (y-ypred)/(1+y) ) )

    # Put the intercept and rest of the coefficients into one array
    c = [res.intercept_]
    [c.append(co) for co in res.coef_]
    c = np.array(c)
    
    # Calculate the least squares error thing (the thing that is actually minimized by the model)
    corr = 0.
    if model == 'ridge':
        corr = alpha * np.sum( (c**2) )
    elif model == 'lasso':
        corr = alpha * np.sum( (np.abs(c)) )
    E_lsq = np.sum( (y - ypred)**2 ) + corr 
    
    # Plotting the residuals vs the redshift
    residuals = ypred - y
    plt.scatter(y, residuals/y, label=str(model), alpha=0.01, marker='.')
    plt.ylabel('(z_spec,pred-z_spec)/z_spec')
    plt.xlabel('z_spec')
    plt.ylim(-0.4, 0.4)
    plt.xlim(0.15,0.7)
    print """Best-fit model is:\n z = {0:.3f}+ {1:.3f} (u-g) + {2:.3f} (g-r) + {3:.3f} (r-i) + {4:.3f} (i-z)""".format(c[0], c[1], c[2], c[3], c[4])
    print "E_theta =", E_theta, 'E_lsq =', E_lsq
    
    #return res
    return model, res, c
```

```python
def make_random_2D_data(N=20, dy = 0.3, dx = 0.3):
    # from Jarle, PS4 solutions
    slope = 2.1
    intercept = 1.4
    xmin = -1.4
    xmax = 4.0
    
    x_true = np.random.uniform(xmin, xmax, size=N)
    y_true = slope*x_true + intercept
    
    x = x_true + np.random.normal(0, dx, size=N)
    y = y_true + np.random.normal(0, dy, size=N)
    
    return x, y

x, y = make_random_2D_data()
```

```python
def find_theta(M, C, Y):
    # This function finds the parameter vector theta (Th)
    # by calculating the solution of linear regression using
    # matrix multiplication
    # M is the design matrix (containing xij)
    # C is the covariance matrix of uncertainties on Y
    # Y are the predictions for the yij as a function of xij
    temp1 = np.matmul(M.transpose(), np.linalg.inv(C))
    temp1 = np.matmul(temp1, M)
    print temp1
    temp1 = np.linalg.inv(temp1)
    print temp1
    temp2 = np.matmul(M.transpose(), np.linalg.inv(C))
    temp2 = np.matmul(temp2, Y)
    Th = np.matmul(temp1, temp2)
    return Th

def find_theta_2(M, Y):
    #return np.matmul(np.linalg.inv(M), Y)
    #print M.shape
    #print Y.shape
    #return np.linalg.solve(M, Y)
    # Doesn't work: also inverses matrix M!
    # use pseudoinverse:
    return np.matmul(np.linalg.pinv(M), Y)

# Y = M*Theta (matrix mul)
# I want to get Theta:
# C = A x B --> B = inv(A) x C
# Theta = inv(M) x Y

# not possible because M is not invertible
# Ax = b solve for X (A = 2x40, x = 2x1, y = 1x40)
# use np.linalg.solve
```

```python
def pca_regression(X, y, model='linear', alpha=0.):
    
    fit_intercept=True
    
    #if normalise == True:
    #    print 'Normalisation not implemented!'
    #    return
    
    # Define pca: without whitening (hard to transform back), and use the max. num of components:
    #pca = PCA(whiten=False) #if n_components is not set all components are kept
    ncomp = len(X[0,:]) + 1
    ncomp = 6
    pca = PCA(whiten=False, n_components = ncomp)

    Data = np.column_stack((X,y)) # combine M and y for input as pca
    
     
    #df = pd.DataFrame(Data)
    #g = sns.pairplot(df)
    
    # Fit data (might be whitened)
    pca.fit(Data) 

    # Transform (rotate) the data and plot the principal components
    PC = pca.transform(Data)
    
    #df2 = pd.DataFrame(PC)
    #g2 = sns.pairplot(df2)
    
    # Regression on the transformed data (implement ridge & lasso later):
    # Use the last PC (least amount of scatter) as Y data
    # and the other PCs as X data
    #ncomp_new = 4
    PC_X = PC[:, :ncomp-1]
    PC_y = PC[:, ncomp-1]
    #print PC_X.shape
    #print PC_y.shape
    
    #print 'first'
    #print PC_X
    #print PC_y
    
    # Try to use the PC with the most scatter as Y data and the other PCs as X data:
    #PC_X = PC[:, 1:len(Data[0,:])]
    #PC_y = PC[:, 0]
    
    # Gives the same result in x, y space, as it should
    # The error is bigger in PC space, so the first approach seems better
    
    #print 'then'
    #print PC_X
    #print PC_y
    
    #PC_mostscatter = PC[:, 0]
    #PC_leastscatter = PC[:, ncomp-1]
    
    if model == 'linear':
        mod = LinearRegression(fit_intercept=fit_intercept) # set fit_intercept to false when using whitened data (?)
    elif model == 'ridge':
        mod = Ridge(alpha, fit_intercept=fit_intercept, normalize=True)
    elif model == 'lasso':
        mod = Lasso(alpha, fit_intercept=fit_intercept, normalize=True)
        
    res = mod.fit(PC_X, PC_y)
    
    #res = mod.fit(PC_mostscatter[:,np.newaxis], PC_leastscatter)
    
    #plt.plot(PC_mostscatter, PC_leastscatter)
    
    # Model: pc_last = c0 + c1*pc1 + c2*pc2 +..., intercept and rest of the coefficients 
    # into one array:
    c_pc = [res.intercept_]
    [c_pc.append(co) for co in res.coef_]
    c_pc = np.array(c_pc)
    
    print 'In PC space, the coefficients are: ', c_pc
    print
    
    # Calculating the predicted values of pc_last given the model (with parameters c) 
    # for given data pc_rest
    ypred_pc = mod.predict(PC_X)
    #ypred_pc = mod.predict(PC_mostscatter[:,np.newaxis])

    # Calculating the error from the problem set in PC space
    E_theta = np.median( np.abs( (PC_y-ypred_pc)/(1+PC_y) ) )
    #E_theta = np.median( np.abs( (PC_leastscatter - ypred_pc)/(1+PC_leastscatter)))
    print 'Etheta for PC fit in PC space:', E_theta
    print
    
    # Find the X and y values of the linear regression model results, but in X, y space
    # by transforming the PC space results to X, y space
    modeldata_pc = np.column_stack((PC_X, ypred_pc))
    Data_model = pca.inverse_transform(modeldata_pc) 

    
    X_model = Data_model[:, :len(X[0,:])]
    y_model = Data_model[:, len(X[0,:])]

    # Solve Y = M x Theta to find Theta for Y and M containing the X_model data
    M_model = np.column_stack((np.ones(len(X_model[:,0])), X_model))
    c_xy = find_theta_2(M_model, y_model) 
                                # this works! #not solved exactly, as the matrix M is not invertible, 
                                # so use pseudo inverse

    print 'In X, y space, the coefficients are: ', c_xy
    print

    # Predicted values for y belonging to the actual X values:
    M = np.column_stack((np.ones(len(X[:,0])), X))
    ypred_fromtheta = np.matmul(M, c_xy) 

    # E_theta for the model in x, y space:
    E_theta = np.median( np.abs( (y-ypred_fromtheta)/(1+y) ) )
    print 'Etheta for X, y fit in X, y space:', E_theta #even a bit lower!
                                        # it turns out that they don't have to be the same!
    print
        
    E_theta_sketchy = np.median( np.abs ( (y - y_model)/(1+y) ) )    
    print 'Sketchy Etheta for model fit in X, y space:', E_theta_sketchy
    print
```
